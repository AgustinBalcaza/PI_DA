```python
import pandas as pd
import numpy as np
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
kpi1 = pd.read_excel('data/homicidios.xlsx', sheet_name='VICTIMAS')
```

```python
kpi1
```

```python
kpi1 = kpi1[['AAAA', 'MM']]
```

```python
kpi1
```

```python
kpi1['SEMESTRE'] = kpi1['MM'].apply(lambda x: 1 if x <= 6 else 2)
```

```python
kpi1
```

```python
kpi1['Victimas'] = 1
df_aggregated = kpi1.groupby(['AAAA', 'SEMESTRE']).agg({'Victimas': 'sum'}).reset_index()
```

```python
kpi1 = df_aggregated
```

```python
kpi1
```

```python
poblacion_values = [3000371, 3010392, 3020412, 3030432, 3040452, 3050472, 3060492, 3070512, 3080532, 3090552, 3100572, 3110592]
```

```python
kpi1['Población'] = poblacion_values
```

```python
kpi1
```

```python
kpi1['Tasa_homicidios'] = (kpi1['Victimas'] / kpi1['Población']) * 100000
```

```python
kpi1
```

```python
kpi1['Variación'] = kpi1['Tasa_homicidios'].pct_change() * 100
```

```python
kpi1
```

```python
kpi1.fillna(0, inplace=True)
```

```python
kpi1
```

```python
kpi1['Variación'] = kpi1['Variación'].round()
```

```python
kpi1
```

```python
kpi1['Variación'] = kpi1['Variación'].astype(int)
```

```python
kpi1
```

```python
string_values = ['2016 - Sem 1', '2016 - Sem 2', '2017 - Sem 1', '2017 - Sem 2', '2018 - Sem 1', '2018 - Sem 2', '2019 - Sem 1', '2019 - Sem 2', '2020 - Sem 1', '2020 - Sem 2', '2021 - Sem 1', '2021 - Sem 2']
```

```python
kpi1['String'] = string_values
```

```python
kpi1
```

```python
kpi1.to_excel('data/KPI1.xlsx', sheet_name='kpi1', index=False)
```

```python
kpi2 = pd.read_excel('data/hechos.xlsx', sheet_name='Sheet1')
```

```python
kpi2
```

```python
kpi2 = kpi2[kpi2['VICTIMA'] == 'MOTO']
```

```python
kpi2
```

```python
kpi2 = kpi2[['AAAA', 'VICTIMA']]
```

```python
kpi2
```

```python
kpi2 = kpi2.groupby('AAAA').size().reset_index(name='Cantidad')
```

```python
kpi2
```

```python
kpi2['Variacion'] = kpi2['Cantidad'].shift(1, fill_value=0)
```

```python
kpi2
```

```python
kpi2['Variacion'] = (kpi2['Variacion'] - kpi2['Cantidad']) / (kpi2['Variacion']) * 100
```

```python
kpi2
```

```python
kpi2['Variacion'] = kpi2['Variacion'].apply(lambda x: -x if x > 0 else abs(x))
```

```python
kpi2
```

```python
kpi2['Variacion'] = kpi2['Variacion'].replace(np.inf, 0)
```

```python
kpi2
```

```python
kpi2['Variacion'] = kpi2['Variacion'].round()
```

```python
kpi2
```

```python
kpi2['Variacion'] = kpi2['Variacion'].astype(int)
```

```python
kpi2
```

```python
kpi2.to_excel('data/KPI2.xlsx', sheet_name='kpi2', index=False)
```
