# Transformaciones

Importaremos la hoja 'HECHOS'

```python
hechos = pd.read_excel('data/homicidios.xlsx', sheet_name='HECHOS')
```

```python
hechos
```

Reemplazamos los nombres de algunas columnas

```python
new_names = {'LUGAR_DEL_HECHO': 'LUGAR', 'TIPO_DE_CALLE': 'TIPO_CALLE', 'Calle': 'CALLE', 'Altura': 'ALTURA', 'Cruce': 'CRUCE', 'Dirección Normalizada': 'DIRECCION', 'XY (CABA)': 'XY_(CABA)', 'pos x': 'POS_X', 'pos y': 'POS_Y'}
```

```python
hechos = hechos.rename(columns=new_names)
```

```python
hechos
```

Obtenemos la información del dataframe, como el tipo de datos, la cantidad de nulos y duplicados con una función del archivo herramientas.py

```python
h.data_info(hechos)
```

Corroboramos por qué en la columna 'HH' están presentes valores de tipo string y qué valores son

```python
valores_string = hechos[hechos['HH'].apply(lambda x: isinstance(x, str))]['HH']

print(valores_string)
```

Como podemos ver, en la fila 518 de la columna 'HH' hay un valor de tipo string y es 'SD', esto está representando que el valor es nulo. Por lo que también rellenaremos el resto de valores nulos con el valor 'SD'

```python
hechos = hechos.fillna('SD')
```

```python
h.data_info(hechos)
```

```python
hechos
```

Ahora cambiaremos los valores flotantes a valores enteros en la columna 'ALTURA' y cambiaremos los valores en 'FECHA' y 'HORA' a tipo string

```python
hechos['ALTURA'] = hechos['ALTURA'].apply(lambda x: int(x) if isinstance(x, (float, int)) else x)
```

```python
hechos['FECHA'] = hechos['FECHA'].astype(str)
```

```python
hechos['HORA'] = hechos['HORA'].astype(str)
```

```python
h.data_info(hechos)
```

```python
hechos
```

Finalmente guardamos los datos transformados en un archivo Excel

```python
hechos.to_excel('data/hechos.xlsx', index=False)
```
